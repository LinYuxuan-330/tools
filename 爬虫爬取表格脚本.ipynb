```python
'''

Created on 2021年2月23日
@author: lyx


'''

#coding=utf-8
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#伪装主机头
head = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

#获取内容
html = requests.get('http://www.xzqh.org/old/waiguo/asia/1001.htm',headers = head)

#转码
html.encoding = 'gb2312'
text = html.text
r = []
soup = BeautifulSoup(text).body

#find_all"tr",align="right"
for i in soup.find("table",border="1").find_all("tr",align="right"):
    
    #获取td即表格中的第一列中的数据
    for j in i.find_all("td")[1].strings:
        r.append(['中国']+[j])
        
#转成DataFrame
a = pd.DataFrame(r)
a
```
